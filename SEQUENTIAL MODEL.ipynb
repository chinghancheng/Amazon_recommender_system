```python
import pandas as pd
import numpy as np
import json
import torch
```

```python
df = pd.read_csv('amazon_reviews_us_Electronics_v1_00.tsv.gz', delimiter='\t',error_bad_lines=False)
df.head()
```

```python
df1 = df.copy()
df1.count()
```

```python
df1 = df1[df1.review_date.notnull()]
df1 = df1[df1.product_title.notnull()]
df1 = df1[df1.product_id.notnull()]
df1 = df1[df1.customer_id.notnull()]
df1.count()
```

```python
def frac(dataframe, fraction, other_info=None):
    """Returns fraction of data"""
    return dataframe.sample(frac=fraction)
df1 = frac(df1, 0.01)
```

```python
df1.count()
```

```python
df1['review_date'] = pd.to_datetime(df1['review_date']).dt.strftime("%Y%m%d")
print(df1.head())
```

```python
# print(df[['product_id']].groupby(df['product_id']).count())
# mean of goup by item
print(df1[['product_id']].groupby(df1['product_id']).count().mean())
# print(df[['customer_id']].groupby(df['customer_id']).count())
# mean of goup by user
print(df1[['customer_id']].groupby(df['customer_id']).count().mean())
print(df1[['star_rating']].astype('int').mean())
```

```python
from collections import defaultdict
from itertools import count

uid_map = defaultdict(count(1).__next__)
iid_map = defaultdict(count(1).__next__)
df1["customer_id"].values

# df1["customer_id"].values
uids = np.array([uid_map[uid] for uid in df1["customer_id"].values ], dtype=np.int32)
iids = np.array([iid_map[iid] for iid in df1["product_id"].values], dtype=np.int32)

uid_rev_map = {v: k for k, v in uid_map.items()}
iid_rev_map = {v: k for k, v in iid_map.items()}

ratings = df1["star_rating"].values.astype(np.float32)
timestamps = df1["review_date"].astype(np.int)

print("userId %d got uid %d" % (41409413, uid_map[41409413]))
print("itemId %s got iid %s" % ('B00428R89M', iid_map['B00428R89M']))
```

```python
print(uids)
print(uid_rev_map)
```

```python
from spotlight.interactions import Interactions
from spotlight.cross_validation import random_train_test_split
from spotlight.cross_validation import user_based_train_test_split

dataset = Interactions(user_ids = uids,
                        item_ids = iids,
                        ratings = ratings,
                        timestamps = timestamps,
                        num_users = len(uid_map)+1,
                        num_items = len(iid_rev_map)+1)

train, test = random_train_test_split(dataset, random_state=np.random.seed(42))
```

```python
print(train)
print(test)
train = train.to_sequence()
test = test.to_sequence()
```

```python
import time
from spotlight.sequence.implicit import ImplicitSequenceModel
start = time.time()
model = ImplicitSequenceModel(n_iter=1,
                              representation='cnn',
                              loss='bpr')
model.fit(train)

end = time.time()
diff = end - start
print('Training time took %d seconds' % (diff))
```

```python
from spotlight.evaluation import sequence_mrr_score
start = time.time()
train_mrr = sequence_mrr_score(model, train)
test_mrr = sequence_mrr_score(model, test)
end = time.time()
diff = end-start
print("MRR on training set",train_mrr,"MRR on testing set",test_mrr,"time-consuming on evaluation %d seconds" % (diff))
```

```python
def sequence_precision_recall_score(model, test, k=10, exclude_preceding=False):
    """
    Compute sequence precision and recall scores. Each sequence
    in test is split into two parts: the first part, containing
    all but the last k elements, is used to predict the last k
    elements.

    Parameters
    ----------

    model: fitted instance of a recommender model
        The model to evaluate.
    test: :class:`spotlight.interactions.SequenceInteractions`
        Test interactions.
    exclude_preceding: boolean, optional
        When true, items already present in the sequence will
        be excluded from evaluation.

    Returns
    -------

    mrr scores: numpy array of shape (num_users,)
        Array of MRR scores for each sequence in test.
    """
    sequences = test.sequences[:, :-k]
    targets = test.sequences[:, -k:]
    precision_recalls = []
    for i in range(len(sequences)):
        predictions = -model.predict(sequences[i])
        if exclude_preceding:
            predictions[sequences[i]] = FLOAT_MAX

        predictions = predictions.argsort()[:k]
        precision_recall = _get_precision_recall(predictions, targets[i], k)
        precision_recalls.append(precision_recall)

    precision = np.array(precision_recalls)[:, 0]
    recall = np.array(precision_recalls)[:, 1]
    return precision, recall
```
